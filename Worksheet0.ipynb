```python
# Task 1: Classify Study Time

# 1. Initialize the study time data
time_data = [1.5, 4.0, 6.5, 2.0, 5.0, 3.5, 7.0, 1.0, 4.5, 5.5]

# 1. Create empty lists for study time classifications
low_study_time = []
moderate_study_time = []
high_study_time = []

# 2. Iterate over the time_data list and add each study hour to the appropriate category
for hours in time_data:
    if hours < 3:
        low_study_time.append(hours)
    elif 3 <= hours <= 5:
        moderate_study_time.append(hours)
    else:
        high_study_time.append(hours)

# 3. Print the lists to verify the classifications
print("Study Time Classification Results")
print(f"Low Study Time (< 3 hours): {low_study_time}")
print(f"Moderate Study Time (3 to 5 hours): {moderate_study_time}")
print(f"High Study Time (> 5 hours): {high_study_time}")

print("\n" + "="*50 + "\n")
```

```python
# Task 2: Based on Data â€“ Answer all the Questions

print("Data Analysis Results ")

# 1. How many days had low study time? (Count items in the low study list)
low_count = len(low_study_time)
print(f"1. Days with Low study time (< 3 hours): {low_count} days")

# 2. How many days had moderate study time? (Count items in the moderate study list)
moderate_count = len(moderate_study_time)
print(f"2. Days with Moderate study time (3 to 5 hours): {moderate_count} days")

# 3. How many days had high study time? (Count items in the high study list)
high_count = len(high_study_time)
print(f"3. Days with High study time (> 5 hours): {high_count} days")

print("\n" + "="*50)
print(f"Total days analyzed: {len(time_data)} days")
```

```python

time_data_hours = [1.5, 4.0, 6.5, 2.0, 5.0, 3.5, 7.0, 1.0, 4.5, 5.5]

## Task 3. Convert Study Hours to Minutes

# 1. and 2. Iterate over the time_data list and store the results
study_minutes = []
for hours in time_data_hours:
    minutes = hours * 60
    study_minutes.append(minutes)

# 3. Print the converted values
print("Task 3: Study Hours Converted to Minutes")
# Display original hours next to minutes for clarity
for i in range(len(time_data_hours)):
    print(f"Original: {time_data_hours[i]} hours | Converted: {study_minutes[i]} minutes")

print("\n" + "="*50 + "\n")


```

```python

## Task 4. Analyze Average Time Use

structured_time_data = [
    (1.5, 3.0, 7.5),
    (4.0, 2.0, 8.0),
    (6.5, 1.5, 6.0),
    (2.0, 4.0, 7.0),
    (5.0, 1.0, 7.0),
    (3.5, 2.5, 8.5)
]

# 1. Create empty lists for study_hours, entertainment_hours, and sleep_hours
study_hours = []
entertainment_hours = []
sleep_hours = []

# 2. Iterate over time_data and extract values into each list
for record in structured_time_data:
    study_hours.append(record[0])
    entertainment_hours.append(record[1])
    sleep_hours.append(record[2])

# Helper function to calculate average
def calculate_average(data_list):
    if not data_list:
        return 0
    return sum(data_list) / len(data_list)

# 3. Calculate and print the averages
print("Task 4: Analyze Average Time Use")

# (a) Average hours spent studying
avg_study = calculate_average(study_hours)
print(f" (a) Average hours spent **studying**: {avg_study:.2f} hours")

# (b) Average hours spent on entertainment
avg_entertainment = calculate_average(entertainment_hours)
print(f" (b) Average hours spent on **entertainment**: {avg_entertainment:.2f} hours")

# (c) Average hours spent sleeping
avg_sleep = calculate_average(sleep_hours)
print(f" (c) Average hours spent **sleeping**: {avg_sleep:.2f} hours")

print("\n" + "="*50 + "\n")
```

```python
## Task 5. Visualization - Study vs Sleep Pattern

# 1. Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt

print("Task 5: Generating Study vs Sleep Scatter Plot")

# 2. Plot a scatter plot
plt.figure(figsize=(8, 6))

plt.scatter(
    study_hours,   # x-axis data
    sleep_hours,   # y-axis data
    color='purple',
    s=100,         # size of markers
    alpha=0.7,     # transparency
    edgecolors='w' # white border for better visibility
)

# 3. Add labels, title, and color
plt.title('Daily Study Hours vs. Sleep Hours Pattern', fontsize=14)
plt.xlabel('Study Hours (x-axis)', fontsize=12)
plt.ylabel('Sleep Hours (y-axis)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Display the plot
plt.show()


print("The scatter plot has been generated and displayed.")
```
